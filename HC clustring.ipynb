```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset
- Reading the dataset csv file 
- : is to get all the rows and [3,4] is that what columns user wants

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

# Using the dendrogram to find the optimal number of clusters

- 'scipy' is a open source python library which contains tools for dendrogram and help to plot dendrogram 
- 'linkage' is itself a algorithm for HC
- 'ward' is method which try to minimize varients in each clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

# Fitting Hierarchical Clustering to the dataset
## importing Agglomerative clustring 
- creating object of class agglomerative than passing the parameters
- fitting HC in our dataset X the fit predict method going to return vector of clusters

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')#ward- minimises the variances within clusters 
y_hc = hc.fit_predict(X)
```

# Visualising the clusters
- in this we find which cluster belongs to which customer. 

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Careful')  
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Standard')  
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Target')  
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Careless')  
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Sensible')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
